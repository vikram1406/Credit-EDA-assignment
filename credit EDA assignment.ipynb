```python
#now can create two seperate value for people who defaulted and who didn't default.
app_0=app_data[app_data.TARGET==0]    
app_1=app_data[app_data.TARGET==1] 
```

```python
# Defining function to count plot for categorical variables
def uninevar(x):

    plt.style.use('ggplot')
    sns.despine
    fig,(ax1,ax2) = plt.subplots(1,2,figsize=(20,5))
    
    sns.countplot(x=x, data=app_0,ax=ax1)
    ax1.set_ylabel('Total Counts')
    ax1.set_title(f'Distribution of {x} for Non-Defaulters',fontsize=20)
    ax1.set_xticklabels(ax1.get_xticklabels(), rotation=50, ha="right")
    
    # Adding the normalized percentage for easier comparision between defaulter and non-defaulter
    for p in ax1.patches:
        ax1.annotate('{:.1f}%'.format((p.get_height()/len(app_0))*100), (p.get_x()+0.1, p.get_height()+70))
        
    sns.countplot(x=x, data=app_1,ax=ax2)
    ax2.set_ylabel('Total Counts')
    ax2.set_title(f'Distribution of {x} for Defaulters',fontsize=20)    
    ax2.set_xticklabels(ax2.get_xticklabels(), rotation=50, ha="right")
    
    for p in ax2.patches:
        ax2.annotate('{:.1f}%'.format((p.get_height()/len(app_1))*100), (p.get_x()+0.1, p.get_height()+50))
    
    plt.show()
```

```python
#plotting defaulters and non defaulters with OCCUPATION_TYPE

uninevar('OCCUPATION_TYPE')
```

--hence we can conclude that there are more number of labours applying for loans and that's why we can see that there are more numbers of defaulters in labours

--we can see that student and businessman don't default as students are not allowed to pay while studying and bussinessman pay loans on time, while there are more number of defaulters than non defaulters in working sector.

```python
#plotting defaulters and non defaulters with CODE_GENDER
uninevar('CODE_GENDER')
```

# Bivariate Analysis

```python
# Difining function for scatter plot for continuous variables
def bivar(x,y):

    plt.style.use('ggplot')
    sns.despine
    fig,(ax1,ax2) = plt.subplots(1,2,figsize=(20,5))
    
    sns.scatterplot(x=x, y=y,data=app_1,ax=ax1)
    ax1.set_xlabel(x)    
    ax1.set_ylabel(y)
    ax1.set_title(f'{x} vs {y} for Non-Defaulters',fontsize=20)
    
    sns.scatterplot(x=x, y=y,data=app_1,ax=ax2)
    ax2.set_xlabel(x)    
    ax2.set_ylabel(y)
    ax2.set_title(f'{x} vs {y} for Defaulters',fontsize=20)
            
    plt.show()
```

```python
corr_0=app_0[['AMT_GOODS_PRICE','AMT_INCOME_TOTAL','AMT_ANNUITY','DAYS_EMPLOYED',
  'DAYS_BIRTH', 'DAYS_REGISTRATION', 'DAYS_ID_PUBLISH',
  'AMT_CREDIT']].corr(method = 'pearson')

corr=corr.where(np.triu(np.ones(corr.shape),k=1).astype(np.bool))
corr_df=corr.unstack().reset_index()

corr_df.columns = ['Col1','Col2','CORRELATION']
corr_df.dropna(subset=['CORRELATION'],inplace=True)
corr_df['CORR_ABS']=corr_df['CORRELATION'].abs()
corr_df.sort_values('CORR_ABS', ascending=False).head(10)
```

```python
corr_1=app_1[['AMT_GOODS_PRICE','AMT_INCOME_TOTAL','AMT_ANNUITY','DAYS_EMPLOYED',
  'DAYS_BIRTH', 'DAYS_REGISTRATION', 'DAYS_ID_PUBLISH',
  'AMT_CREDIT']].corr(method = 'pearson')

corr=corr.where(np.triu(np.ones(corr.shape),k=1).astype(np.bool))
corr_df=corr.unstack().reset_index()

corr_df.columns = ['Col1','Col2','CORRELATION']
corr_df.dropna(subset=['CORRELATION'],inplace=True)
corr_df['CORR_ABS']=corr_df['CORRELATION'].abs()
corr_df.sort_values('CORR_ABS', ascending=False).head(10)
```

```python
#plotting scatter plot for 'AMT_CREDIT','AMT_GOODS_PRICE'.
bivar('AMT_CREDIT','AMT_GOODS_PRICE')
```

--As there is a linear relation between AMT_GOODS_PRICE vs AMT_CREDIT as they are increasing linearly

```python
#bivirate analysis function
def catnum(cat, num):

    plt.style.use('ggplot')
    sns.despine
    fig,ax = plt.subplots(1,1,figsize=(12,8))
    
    sns.boxenplot(x=cat,y = num, data=pre_data)
    ax.set_ylabel(f'{num}')
    ax.set_xlabel(f'{cat}')

    ax.set_title(f'{cat} Vs {num}',fontsize=15)
    ax.set_xticklabels(ax.get_xticklabels(), rotation=50)
     
    plt.show()
```

# Merging both the dataset for analysis

```python
merge_data= pd.merge(app_data, pre_data, how='inner', on=['SK_ID_CURR'])
merge_data.head()
```

```python
merge_data.info()
```

```python
def mergeuni(Varx,Vary):
    
    plt.style.use('ggplot')
    sns.despine
    mergenew= merge_data.pivot_table(values='SK_ID_CURR', 
                      index=Varx,
                      columns=Vary,
                      aggfunc='count')
    mergenew=mergenew.div(mergenew.sum(axis=1),axis='rows')*100
    sns.set()
    mergenew.plot(kind='bar',stacked=True,figsize=(20,5))
    plt.title(f'Effect Of {Varx} on Loan Approval')
    plt.xlabel(f'{Varx}')
    plt.ylabel(f'{Vary}%')
    plt.show()
```

```python
#Univariate analysis for merge data
mergeuni('TARGET','NAME_CONTRACT_STATUS')
```

--people who were approved for a loan earlier defaulted less where people who were refused a loan earlier have higher chances of defaulting.
